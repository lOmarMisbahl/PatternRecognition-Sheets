```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
import numpy as np
import seaborn as sns
```

```python
iris = datasets.load_iris()
labels = iris.target
```

```python
colors = ['blue', 'red','green']
plt.figure(figsize=(5,5))
for label, color in zip(range(len(labels)), colors):
    plt.hist(iris.data[labels==label, 0],
              label=np.unique(labels)[label],
              color=color)
plt.show()
```

```python
fig = plt.figure(figsize = (16, 9))
ax = plt.axes(projection ="3d")
   
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.3,
        alpha = 0.2)
 
my_cmap = plt.get_cmap('hsv')

sctt = ax.scatter3D(iris.data[:,0], iris.data[:,1], iris.data[:,2],
                    alpha = 1,
                    c = (iris.data[:,0] + iris.data[:,1] + iris.data[:,2]),
                    cmap = my_cmap,
                    marker ='^')
 
ax.set_xlabel(iris.feature_names[0], fontweight ='bold')
ax.set_ylabel(iris.feature_names[1], fontweight ='bold')
ax.set_zlabel(iris.feature_names[2], fontweight ='bold')
fig.colorbar(sctt, ax = ax, shrink = 0.5, aspect = 5)
 
plt.show()
```
