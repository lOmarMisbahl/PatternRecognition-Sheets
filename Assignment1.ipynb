```python
import matplotlib.pyplot as plt
import math as mth
import plotly.express as px
import numpy as np
from sklearn.metrics import pairwise_distances
from sklearn.metrics.pairwise import euclidean_distances
```

```python
d2 = np.matrix([[10, 60, 10, 90], [20, 50, 40, 70], [30, 50, 30, 40], [20, 50, 20, 60], [10, 60, 30, 10]])
print("Data:\n",d2)
```

```python
norm = np.linalg.norm(d2, axis=1)
print("Norm:\n",norm)
```

```python
CosSimi = 1-pairwise_distances(d2, metric="cosine")
print("Cosine Similarity:\n",CosSimi)
```

```python
EuclDist= euclidean_distances(d2,d2)
print("Euclidean Distance:\n",EuclDist)
```

```python
d3 = np.matrix([[10, 60, 90], [20, 50, 70], [30, 50, 40], [20, 50, 60], [10, 60, 10]])
print("Data:\n",d3)
```

```python
fig = plt.figure(figsize = (16, 9))
ax = plt.axes(projection ="3d")
   
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.3,
        alpha = 0.2)
 
my_cmap = plt.get_cmap('hsv')

sctt = ax.scatter3D(d3[:,0], d3[:,1], d3[:,2],
                    alpha = 1,
                    c = (d3[:,0] + d3[:,1] + d3[:,2]),
                    cmap = my_cmap,
                    marker ='^')
 
ax.set_xlabel("a1", fontweight ='bold')
ax.set_ylabel("a2", fontweight ='bold')
ax.set_zlabel("a4", fontweight ='bold')
fig.colorbar(sctt, ax = ax, shrink = 0.5, aspect = 5)
 
plt.show()
```

```python
mean= np.mean(d3, axis=0)
print("Mean:\n",mean)
```

```python
Z = d3 - mean
print("Centered Data:\n",Z)
```

```python
cov = (1/5)*np.dot(np.transpose(Z), Z)
print("Covariance Matrix:\n",cov)
```

```python
V, U = np.linalg.eig(cov)
print("Eigen Values:\n",V)
print("Eigen Vectors:\n",U)
```

```python
DiagEigenvalues= np.diag(V)
CovVerify = np.dot(np.dot(U, DiagEigenvalues), np.transpose(U))
print("Covariance Verification:\n",CovVerify)
```

```python
LargestEigenValue = V.max()
ExplainedVariance = LargestEigenValue/np.sum(V)
print("Explained Variance:",ExplainedVariance)
```

```python
Pairs = [(np.abs(V[i]), U[:,i]) for i in range(len(V))]
Pairs.sort(key=lambda x: x[0], reverse=True)
Projection = np.hstack((Pairs[0][1].reshape(3,1), Pairs[1][1].reshape(3,1)))
print("Projection Matrix:\n",Projection)
```

```python
Projected= np.dot(d3, Projection)
print("Projected Data Matrix:\n",Projected)
```

```python
fig= px.scatter_matrix(Projected)
fig.show()
```
